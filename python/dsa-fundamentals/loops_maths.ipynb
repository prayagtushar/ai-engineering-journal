### While Loop

```python
n = 1
while n < 10:
    n = n + 5
    print(n)

# Output: 6, 11
```

### For Loop

```python
# Basic: 0 to 4
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()
```

```python
# With start and end: 2 to 5 (exclusive end)
print("range(2,6):")
for i in range(2,6):
    print(i, end=" ")
print()
```

```python
# With step: from 10, decreasing by 2 until lowest number before 2
print("range(10, 4, -2):")
for i in range(10, 4, -2):
    print(i, end=" ")
print()
# Output: 10, 8, 6
```

### Division Operators

```python
# / gives float result
print("5 / 2 =", 5 / 2)    # 2.5

# // gives integer result (floor division)
print("5 // 2 =", 5 // 2)  # 2
```

### Negative Integer Division

```python
# By default, negative integers don't round down
print("-3 / 2 =", -3/2)  # -1.5

# Use int() to round down to closest integer
print("int(-3/2) =", int(-3/2))  # -1
```

```python
# Remainder operator
print("10 % 3 =", 10 % 3)  # 1
```

### Negative Remainder

```python
# For negative numbers, % works differently than expected
print("-10 % 3 =", -10 % 3)

# Workaround: use math.fmod
import math
a = math.fmod(-10, 3)
print("math.fmod(-10, 3) =", a)
```

### Math Functions

```python
num = 3/2
print("num =", num)
print("floor:", math.floor(num))   # rounds down
print("ceil:", math.ceil(num))    # rounds up
print("sqrt:", math.sqrt(num))    # square root
print("pow:", math.pow(num, 2))   # power
```

### Infinity

```python
# Represents highest/lowest possible numbers
print("Positive infinity:", float("inf"))
print("Negative infinity:", float("-inf"))
```
