### HashSet

A collection with **no duplicates** - O(1) lookup time

```python
# Create empty set
myset = set()
myset.add(1)
myset.add(2)
myset.add(3)
myset.add(5)
print("Set:", myset)
```

```python
# Length
print("Length:", len(myset))
```

```python
# Search in O(1) - no function needed
print("2 in myset:", 2 in myset)  # True
```

```python
# Remove by value (not index)
myset.remove(3)
print("After removing 3:", myset)  # {1, 2, 5}
```

### Initialize with values

```python
# From list
myset = set([1,2,3,4,5,6])
print("From list:", myset)
```

```python
# Using set comprehension (like list but with {})
myset = {i for i in range(5)}
print("Set comprehension:", myset)  # {0, 1, 2, 3, 4}
```

### HashMap (Dictionary)

Key-value pairs with **no duplicate keys** - O(1) lookup time

```python
# Create empty dict
m = {}  

# Easy to assign value
m["a"] = 1000
m["z"] = 1

print("Dict:", m)  # {'a': 1000, 'z': 1}

# Mutable - can change value
m["a"] = "zero"
print("After update:", m)  # {'a': 'zero', 'z': 1}
```

```python
# Search using 'in' - returns True/False
print("'a' in m:", "a" in m)  # True
```

```python
# Remove key
m.pop('a')
print("After pop:", m)
```

### Initialize with values

```python
m = {'a': 120, 'b': 500}
# 'a' and 'b' are keys
# 120 and 500 are values
print(m)
```

```python
# Using dict comprehension
m = {i: 2 * i for i in range(5)}
print("Dict comprehension:", m)  # {0: 0, 1: 2, 2: 4, 3: 6, 4: 8}
```

### Looping through HashMap

```python
m = {i: 2 * i for i in range(2)}

# Method 1: Loop through keys
print("Method 1 - Keys:")
for key in m:
    print(f"Key: {key}, Value: {m[key]}")
```

```python
# Method 2: Loop through values only
print("Method 2 - Values:")
for value in m.values():
    print(f"Value: {value}")
```

```python
# Method 3: Loop through key-value pairs (most useful)
print("Method 3 - Items:")
for key, val in m.items():
    print(f"Key: {key}, Value: {val}")
```
