### Variables

Python is **dynamically typed** - no type declaration needed

```python
n = 10
print("Integer:", n)  # 10

# Can reassign to different type
n = "ambulance"
print("String:", n)  # ambulance
```

### Important Notes

```python
# n++ doesn't work in python
# Use n += 1 instead

# None is Python's null value
x = None
print("None:", x)
```

### If Statement

**No parentheses or curly braces required** - indentation matters!

```python
n = 10
if n > 10:
    n = n - 1
elif n < 10:  # "else if" is "elif" in Python
    n = n + 2
else:
    print("n is 10")

print("Result:", n)
```

### Multiple Conditions

- `&&` → `and`
- `||` → `or`
- Use parentheses for complex conditions

```python
n = 10
if ((n > 2) and n % 2 == 0) or (n == 10):
    print("COMPLETE")
```
