### Nested Functions

Functions defined inside another function

```python
def nested(a,b,c):
    sum = a+b

    def inner():
        finalSum = sum + c
        return finalSum
    return inner()

print(nested(1,5,10))
```

### Modifying Variables in Nested Functions

**Rare use case:** Functions can modify objects but cannot reassign unless using `nonlocal` keyword

```python
def doubleValue(arr, val):
    
    def helperFn():
        # This modifies the list in-place (works fine)
        for i in range(len(arr)):
            arr[i] *= 2

        # Uncommenting would only modify val in helper scope
        # val *= 2

        # To modify outer scope variable, use 'nonlocal'
        nonlocal val
        val *= 2
    
    helperFn()
    print(arr, val)

nums = [1,2]
val = 2
doubleValue(nums, val)  # Output: [2, 4] 4
```

### Classes

```python
class cls:
    # Constructor - self is like 'this' in other languages
    def __init__(self, nums):
        self.nums = nums
        self.size = len(nums)

    # Self needs to be passed for member functions
    def getL(self):
        return self.size
    
    def getDoubleLength(self):
        return 2 * self.getL()

# Create instance
obj = cls([1,2,3])
print("Length:", obj.getL())
print("Double Length:", obj.getDoubleLength())
```
