# Pytest Foundations: Testing Your Code Like a Pro

Testing isn't just about finding bugs; it's about having the confidence to change your code without breaking it. 

### Why Pytest?
- **Simple**: Tests are just functions.
- **Powerful**: Features like fixtures and parametrization make complex tests easy.
- **Industry Standard**: Almost every Python AI project uses Pytest.

## 1. The Basics: How Pytest Finds Tests

Pytest is smart. It looks for files and functions with specific names:
- **Files**: `test_*.py` or `*_test.py`
- **Functions**: `test_*` 

**Beginner Tip**: An "Assertion" is just a check. If `assert 1 + 1 == 2` is true, the test passes. If it's false, the test fails.

```python
def add(a, b):
    return a + b

def test_add():
    # This is a basic test case
    assert add(1, 2) == 3
    assert add(-1, 1) == 0
```

## 2. Fixtures: The "Setup" Helpers

Imagine you need a database connection or a specific piece of data for 10 different tests. Instead of creating it in every test, you create a **Fixture**.

Think of a fixture as a "prepared ingredient" that you can inject into any test function.

```python
import pytest

# This fixture provides a sample user dictionary
@pytest.fixture
def sample_user():
    return {"username": "dev_hero", "level": 99}

def test_user_level(sample_user):
    # We just 'ask' for the fixture by name in the function arguments
    assert sample_user["level"] > 50
```

## 3. Parametrization: One Test, Many Inputs

Instead of writing a new test function for every scenario, use `@pytest.mark.parametrize` to run the same test with different data.

**Why this matters**: It keeps your test code DRY (Don't Repeat Yourself) and makes it easy to add new test cases.

```python
@pytest.mark.parametrize("a, b, expected", [
    (1, 1, 2),   # Simple addition
    (0, 5, 5),   # Zero
    (-1, -1, -2) # Negative numbers
])
def test_add_variations(a, b, expected):
    assert add(a, b) == expected
```

## 4. Async Tests: Testing the Future

In AI projects, almost everything is `async` (LLM calls, database queries). To test an `async def` function, you need the `pytest-asyncio` plugin.

**Beginner Tip**: Don't forget the `@pytest.mark.asyncio` decorator, otherwise the test won't run correctly!

```python
import asyncio
import pytest

async def async_fetch_data():
    await asyncio.sleep(0.1) # Simulate a network call
    return "Secret Data"

@pytest.mark.asyncio
async def test_async_fetch():
    result = await async_fetch_data()
    assert result == "Secret Data"
```

## 5. Mocking: The Art of Faking It

What if you want to test code that calls a real API (like OpenAI) but you don't want to spend money or wait for the network? You **Mock** it.

Mocking means replacing a real component with a fake one that returns a predefined value.

**Why this matters**: It makes your tests **fast**, **cheap**, and **reliable** (no network failures!).

```python
from unittest.mock import MagicMock

def get_ai_prediction(client, input_text):
    # This would normally call a real AI model
    return client.predict(input_text)

def test_mock_ai():
    # 1. Create a fake client
    mock_client = MagicMock()
    
    # 2. Tell the fake client what to return
    mock_client.predict.return_value = "I am a fake AI response"
    
    # 3. Call your function with the fake client
    result = get_ai_prediction(mock_client, "Hello!")
    
    # 4. Check if it worked and if the fake client was called correctly
    assert result == "I am a fake AI response"
    mock_client.predict.assert_called_with("Hello!")
```
