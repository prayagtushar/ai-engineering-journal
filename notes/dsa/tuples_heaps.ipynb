### Tuples

Like arrays but **immutable** - cannot be changed after creation

```python
arr = (1, 2, 4)
print("Tuple:", arr)
```

```python
# Uncomment to see error - tuples are immutable
# arr[0] = 120  # TypeError
```

### Why use Tuples?

Tuples can be used as **keys in dictionaries** and added to **sets** - lists cannot do this!

```python
# Tuple as dictionary key
myhash = {(1,2): 5}
print("Dict with tuple key:", myhash)

# Tuple in set
s = set()
s.add((1,2))
print("Set with tuple:", s)
print("(1,2) in s:", (1,2) in s)
```

### Heaps

Priority queue implementation - always keeps **minimum element at index 0**

```python
import heapq

minHeap = []
heapq.heappush(minHeap, 2)
heapq.heappush(minHeap, 1232312)
heapq.heappush(minHeap, 123)
heapq.heappush(minHeap, 1)
print("Min Heap:", minHeap)  # [1, 2, 123, 1232312]
```

By default, Python heaps are **min heaps** - minimum is always at index 0

```python
# Pop all elements (sorted order)
print("Popping all:")
while minHeap:
    print(heapq.heappop(minHeap), end=" ")
print()
```

### Max Heap

Python doesn't have max heap by default. **Workaround:** multiply values by -1

```python
maxH = []
heapq.heappush(maxH, -3)
heapq.heappush(maxH, -2)
heapq.heappush(maxH, -102)
heapq.heappush(maxH, -300)
print("Max Heap (stored as negatives):", maxH)

# Get max value
print("Max value:", -1 * maxH[0])

# Pop all in max order
print("Popping all (max first):")
while maxH:
    print(-1 * heapq.heappop(maxH), end=" ")
print()
# Output: 300, 102, 3, 2
```

### Heapify

Build a heap from existing list in O(n) time

```python
arr = [1000, 2, 3, 45]
heapq.heapify(arr)
print("Heapified:", arr)  # [2, 45, 3, 1000]
```
