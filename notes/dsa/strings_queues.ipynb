### Strings

Strings are **immutable** - cannot be modified in place

```python
s = "aaaaa"
print("Original:", s)
print("Slicing s[0:2]:", s[0:2])
```

```python
# Concatenation creates a new string
s += "bbbb"
print("After +=:", s)
```

### Type Conversion

```python
# String to int (only works for valid numbers)
print("int('123') + int('122') =", int("123") + int("122"))

# Int to string
print("str(11123) + str(1000) =", str(11123) + str(1000))
```

### ASCII Values

```python
# ord() - character to ASCII
print("ord('a') =", ord('a'))

# chr() - ASCII to character
print("chr(97) =", chr(97))
```

### Join Strings

```python
arr = ['a', 'b', 'c']

# Join with empty string
print("Join with '':", ''.join(arr))

# Join with delimiter
print("Join with '-':", '-'.join(arr))
```

### Queue (using deque)

Double-ended queue - all operations are **O(1)**

```python
from collections import deque

que = deque()
que.append(1)
que.append(2)
que.append(3)
que.append(4)
que.append(5)
print("Queue:", que)  # deque([1, 2, 3, 4, 5])
```

```python
# Add to front
que.appendleft(100)
print("After appendleft(100):", que)  # deque([100, 1, 2, 3, 4, 5])
```

```python
# Remove from end
print("pop():", que.pop())  # 5

# Remove from front
print("popleft():", que.popleft())  # 100

print("Final queue:", que)
```

**Note:** All queue operations happen in O(1) time - very efficient!
